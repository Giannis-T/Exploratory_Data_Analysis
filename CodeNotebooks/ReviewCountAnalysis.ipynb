 Εδω παλι βλεπουμε οτι τα περισσοτερα εστιατορια εχουν λιγα reviews, αλλα μπορουμε να δουμε και τις περιπτωσεις με τα περισσοτερα reviews.
### Α.3)

```python
count, bin_edges = np.histogram(df_bus.review_count, bins=100)

c = plt.plot(bin_edges[0:-1], count)
plt.loglog()
```

### A.4)

```python
d = plt.scatter(df_bus.review_count.rank(), df_bus.review_count)
plt.loglog()
```

Βλεπουμε οτι οι περισσοτερες τιμες εχουν υψηλο ranking και λιγες ειναι με χαμηλο, που σημαινει οτι τα περισσοτερα reviews πανε σε λιγες εταιριες(αυτες με το υψηλο ranking) ενω λιγοτερα reviews πανε στις περισσοτερες(αυτες με το χαμηλοτερο ranking).

```python
plt.figure(figsize=(13, 13))

plt.subplot(221)
a = df_bus.review_count.hist(bins=100)
plt.title("Histogram review_count")


plt.subplot(222)
df_bus.review_count.hist(bins=100, log=True)
plt.title("Histogram review_count log-log scale")

plt.subplot(223)
c = plt.plot(bin_edges[0:-1], count)
plt.loglog()
plt.title("cdf of review_count")

plt.subplot(224)
d = plt.scatter(df_bus.review_count.rank(), df_bus.review_count)
plt.loglog()
plt.title("Zipf distribution of review_count")
```

Έχοντας παρατηρήσει τις παραπάνω γραφικές συμπεραίνουμε ότι τα περισσότερα reviews γίνονται σε λίγα καταστήματα, ενώ τα περισσότερα καταστήματα έχουν λίγα reviews.
